```python
x = np.arange(1, 101, 1)
F = 2 * x ** 2 + 5
table = np.column_stack((x, F))
print(table)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

***Data distribution***

```python
fig, axs = plt.subplots(ncols=3)
fig.set_size_inches(16, 5)

sns.histplot(penguins['bill_length_mm'], bins=10, ax=axs[0])
sns.kdeplot(penguins['bill_length_mm'], fill=True, bw_adjust=.5, cut=0, ax=axs[1])
_ = sns.ecdfplot(penguins['bill_length_mm'], ax=axs[2])
```

***Regression***

```python
_ = sns.regplot(x="bill_depth_mm", y="bill_length_mm", data=penguins)
```

***Scatterplots***

```python
_ = sns.catplot(y="body_mass_g", x="flipper_length_mm",  data=penguins, kind='strip')
```

```python
_ = sns.catplot(y="body_mass_g", x="flipper_length_mm", kind="swarm", data=penguins)
```

***Distribution plots***

```python
_ = sns.catplot(x="body_mass_g", y="flipper_length_mm", hue="sex", kind="box", data=penguins)
```

```python
_ = sns.catplot(x="body_mass_g", y="flipper_length_mm", hue="sex", kind="violin", orient="h", data=penguins)
```
